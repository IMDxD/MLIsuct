```python
def sum_of_x_global(a):
    return x + a
```

```python
sum_of_x_global(5)
```

```python
def sum_of_x_local(a):
    x = 3
    return x + a
```

```python
sum_of_x_local(5)
```

```python
def sum_of_x_global_with_link(a):
    global x
    x = 3
    return x + a
```

```python
sum_of_x_global_with_link(5)
```

```python
x
```

```python
def sum_of_x_nonlocal(a):
    x = 3
    def wrapper():
        nonlocal x
        x = 4
        return x
    return wrapper() + a
```

```python
sum_of_x_nonlocal(5)
```
