# Матричные производные

## Производная

Вспомним определение производной:

$$\large{f'(x) = \lim_{h \to 0}\frac{f(x + h) - f(x)}{h}}$$


## Частная производная
Для частных производных h применяется для выбраного базиса

$$\large{\frac{\partial f(x)}{\partial x_i} = \lim_{h \to 0}\frac{f(x + he^i) - f(x)}{h}}$$

Где $e^i$ базис для $x_i$

## Градиент
Тогда градиент в точке x равен вектору частных производных функции

$$
\large{g = \frac{\partial f(x)}{\partial x} = \nabla f = 
\begin{pmatrix} 
    \frac{\partial f(x)}{\partial x_1} \\ 
    \frac{\partial f(x)}{\partial x_2} \\ 
    \vdots \\ 
    \frac{\partial f(x)}{\partial x_m} 
\end{pmatrix}}
$$

## Производная по направлению

Производная по направлению измеряет как сильно функция $f$ изменяется вдоль выбраного направления $v$

$$\large{D_{v}f = \lim_{h \to 0}\frac{f(x + h v) - f(x)}{h}}$$

## Якобиан

Пусть имеется функция $f$ которая преобразует вектор размера n в вектор размера m $f : R^n \rightarrow R^m$, тогда Якобиан этой функции:

$$
\large{J_{f}(x) = \frac{\partial f}{\partial x^T} = 
\begin{bmatrix} 
    \frac{\partial f_1}{\partial x_1} \ldots \frac{\partial f_1}{\partial x_n} \\ 
    \vdots \ddots \vdots \\ 
    \frac{\partial f_m}{\partial x_1} \ldots \frac{\partial f_m}{\partial x_n} 
\end{bmatrix}} = 
\begin{pmatrix} 
    \nabla f_1(x)^T \\
    \vdots \\
    \nabla f_m(x)^T
\end{pmatrix} 
$$

### Умножения Якобиана на вектор

Умножение Якобиана на вектор

$$
\large{J_{f}(x)v = 
\begin{pmatrix} 
    \nabla f_1(x)^T \\
    \vdots \\
    \nabla f_m(x)^T
\end{pmatrix} v =
\begin{pmatrix} 
    \nabla f_1(x)^T v \\
    \vdots \\
    \nabla f_m(x)^T v
\end{pmatrix}}
$$
Умножение вектора на Якобиан
$$
\large{u^TJ_{f}(x) = u^T(\nabla f_1(x)^T \ldots \nabla f_m(x)^T) = (u^T \nabla f_1(x)^T \ldots u^T \nabla f_m(x)^T)}
$$

### Композиция Якобиана

$$\large{J_h(x) = J_g(f(x))J_f(x)}$$

## Табличные градиенты

- a, b, x - вектора
- A - матрица

$$\large{\frac{\partial a^T x}{\partial x} = a}$$
$$\large{\frac{\partial b^T A x}{\partial x} = A^T b}$$
$$\large{\frac{\partial x^T A x}{\partial x} = (A + A^T) x}$$

- a, b - вектора
- A, B, X - матрицы

$$
\large{
    \frac{\partial f}{\partial X} = 
    \begin{pmatrix}
        \frac{\partial f_1}{\partial x_{11}} \ldots \frac{\partial f_1}{\partial x_{1n}} \\
        \vdots \ddots \vdots \\
        \frac{\partial f_m}{\partial x_{m1}} \ldots \frac{\partial f_m}{\partial x_{mn}} \\
    \end{pmatrix}
}
$$

$$\large{\frac{\partial a^TXb}{\partial X} = a^Tb}$$
$$\large{\frac{\partial a^TX^Tb}{\partial X} = ba^T}$$
$$\large{\frac{\partial X^TX}{\partial X} = det(AXB)X^{-T}}$$
